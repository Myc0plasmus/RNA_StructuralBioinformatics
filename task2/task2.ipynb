# INF

````python
from math import sqrt

def DotBracketToBp(dbSeq):
    bp = []

    openingBrackets = ["(", "{", "[", "<", "A", "B"]
    closingBrackets = [")", "}", "]", ">", "a", "b"]
    stacks = [ [] for _ in openingBrackets ]

    for i in range(len(dbSeq)):
        if dbSeq[i] in openingBrackets:
            stackIdx = openingBrackets.index(dbSeq[i])
            stacks[stackIdx].append(i)
        elif dbSeq[i] in closingBrackets:
            stackIdx = closingBrackets.index(dbSeq[i])
            j = stacks[stackIdx].pop()
            bp.append((j+1, i+1))
        
    return bp

def calculateINF(dbSeqA, dbSeqB):
    bpA = DotBracketToBp(dbSeqA)
    bpB = DotBracketToBp(dbSeqB)

    setA = set(bpA)
    setB = set(bpB)

    TPset = setA & setB
    
    FNset = setA - setB
    
    FPset = setB - setA

    TP = len(TPset)
    FN = len(FNset)
    FP = len(FPset)

    return sqrt((TP/(TP+FN))*(TP/(TP+FP)))


turboFold = "((..(((((((......((.((((((((...............))))))))..))..)))))))..))"
locARNA = "((..(((((((.........((((.(((...............))).))))......)))))))..))"
locARNAConsensus = "((.((((((((...(.(...((((.(((.(((((.....))))))))...))))).)...)))))))).))"
RFAMConsensus =  "....((((((..........(((((((.....))))))).......))))))...."

print(f"test passed: {abs(calculateINF("(.(((...))).)", "[((((..])))).")-0.67) <  0.01}")
print(f"turboFold & locARNA: {calculateINF(turboFold, locARNA)}")
print(f"locARNA & locARNAConsensus: {calculateINF(locARNA, locARNAConsensus)}")
print(f"turboFold & locARNAConsensus: {calculateINF(turboFold, locARNAConsensus)}")
print(f"RFAMConsensus & locARNAConsensus: {calculateINF(RFAMConsensus, locARNAConsensus)}")
print(f"turboFold & RFAMConsensus: {calculateINF(turboFold, RFAMConsensus)}")
````

# Levenstein

````python
import numpy as np

def LevenshteinDistance(s, t):
    M = len(s)
    N = len(t)
    d = np.zeros([M, N])

    for i in range(0,M):
        d[i][0] = i
    for j in range(0,N):
        d[0][j] = j
    for i in range(1,M):
        for j in range(1,N):
            if s[i] == t[j]:
                cost = 0
            else:
                cost = 1
            d[i][j] = min(d[i-1][j] + 1,
                    d[i][j-1] + 1,
                    d[i-1][j-1] + cost)
    return d[M-1][N-1]

print(f"test passed: {LevenshteinDistance("...(((..)))", "..(((...)))")==2.0}")
print(f"turboFold & locARNA: {LevenshteinDistance(turboFold, locARNA)}")
print(f"locARNA & locARNAConsensus: {LevenshteinDistance(locARNA, locARNAConsensus)}")
print(f"turboFold & locARNAConsensus: {LevenshteinDistance(turboFold, locARNAConsensus)}")
print(f"RFAMConsensus & locARNAConsensus: {LevenshteinDistance(RFAMConsensus, locARNAConsensus)}")
print(f"turboFold & RFAMConsensus: {LevenshteinDistance(turboFold, RFAMConsensus)}")
````

````python
def DotBracketToPairs(dbSeq):
    bp = []

    openingBrackets = ["(", "{", "[", "<", "A", "B"]
    closingBrackets = [")", "}", "]", ">", "a", "b"]
    stacks = [ [] for _ in openingBrackets ]

    for i in range(len(dbSeq)):
        if dbSeq[i] in openingBrackets:
            stackIdx = openingBrackets.index(dbSeq[i])
            stacks[stackIdx].append(i)
        elif dbSeq[i] in closingBrackets:
            stackIdx = closingBrackets.index(dbSeq[i])
            j = stacks[stackIdx].pop()
            bp.append((j+1, i+1))
        else:
            bp.append((i+1,i+1))
    return bp

class treeNode:
    def __init__(self, nodeIdx, level, children):
        self.idx = nodeIdx
        self.level = level
        self.children = children

def printTree(root: treeNode, isRoot=True):
    if isRoot == True:
        return "".join([ printTree(child, False) for child in root.children]) 
    if root.children == []:
        return f"(U{root.level})"
    else:
        return "(" + "".join([ printTree(child, False) for child in root.children]) + f"P{root.level}" +")"
        
    

def makeHitTree(dbSeq):
    pairs = DotBracketToPairs(dbSeq)
    pairs = sorted(pairs, key=lambda x: x[0])

    tree = treeNode(0,0,[])
    boundList = []
    currentNode = tree

    for pairIdx, pair in enumerate(pairs):
        if pair[0] != pair[1]:
            if pairIdx == 0 or pairs[pairIdx-1][1] - 1 != pair[1]:
                if len(boundList) == 0 :
                    tree.children.append(treeNode(pairIdx+1,1,[]))
                    currentNode = tree.children[-1]
                else:
                    boundIdx = -1
                    while(boundIdx >= -len(boundList) and pair[1] > boundList[boundIdx][1][1]):
                        boundIdx -= 1
                    if boundIdx < -len(boundList):
                        tree.children.append(treeNode(pairIdx+1, 1, []))
                        currentNode = tree.children[-1]
                    else: 
                        boundList[boundIdx][2].children.append(treeNode(pairIdx+1, 1, []))
                        currentNode = boundList[boundIdx][2].children[-1]
                        
                boundList.append((pairIdx,pair, currentNode))
            else:
                currentNode.level += 1
        else:
            if pairIdx == 0 or pairs[pairIdx-1][0] != pairs[pairIdx-1][1] or pair[0] - 1 != pairs[pairIdx-1][0]:
                if len(boundList) == 0 :
                    tree.children.append(treeNode(pairIdx+1,1,[]))
                    currentNode = tree.children[-1]
                else:
                    boundIdx = -1
                    while(boundIdx >= -len(boundList) and pair[1] > boundList[boundIdx][1][1]):
                        boundIdx -= 1
                    if boundIdx < -len(boundList):
                        tree.children.append(treeNode(pairIdx+1, 1, []))
                        currentNode = tree.children[-1]
                    else: 
                        boundList[boundIdx][2].children.append(treeNode(pairIdx+1, 1, []))
                        currentNode = boundList[boundIdx][2].children[-1]
            else:
                currentNode.level += 1
    return tree

def transformToHIT(dbSeq):
    return printTree(makeHitTree(dbSeq))

testDB = "(((((((..((((....)))).((((...)))).....(((((....)))))))))))).."
correct = "((U2)((U4)P4)(U1)((U3)P4)(U5)((U4)P5)P7)(U2)"

print(f"test passed: {transformToHIT(testDB) == correct}")
print(f"turboFold & locARNA: {LevenshteinDistance(transformToHIT(turboFold), transformToHIT(locARNA))}")
print(f"locARNA & locARNAConsensus: {LevenshteinDistance(transformToHIT(locARNA), transformToHIT(locARNAConsensus))}")
print(f"turboFold & locARNAConsensus: {LevenshteinDistance(transformToHIT(turboFold), transformToHIT(locARNAConsensus))}")
print(f"RFAMConsensus & locARNAConsensus: {LevenshteinDistance(transformToHIT(RFAMConsensus), transformToHIT(locARNAConsensus))}")
print(f"turboFold & RFAMConsensus: {LevenshteinDistance(transformToHIT(turboFold), transformToHIT(RFAMConsensus))}")
````
